## Average -> Remove Background

```python
def Ave_RemoveBG(df):
    returnval = df.groupby('Time [min]').mean()
    returnval = returnval - returnval.loc['B']
    returnval = returnval.drop('B')
    return returnval
```

```python
def Ave_RemoveBG(df):
    a = df.groupby('Time [min]').mean()
    a.loc[['++1','+-1','-+1','--1'], :] = a.loc[['++1','+-1','-+1','--1'], :] - a.loc['B1']
    a.loc[['++2','+-2','-+2','--2'], :] = a.loc[['++2','+-2','-+2','--2'], :] - a.loc['B2']
    a = a.drop('B1')
    a = a.drop('B2')
    return a
```

```python
yfp_bg = Ave_RemoveBG(yfp)
mkate_bg = Ave_RemoveBG(mkate)
od_bg = Ave_RemoveBG(od)
```

## Smooth with uniform_filter1d from Scipy
https://stackoverflow.com/questions/13728392/moving-average-or-running-mean/43200476#43200476

```python
from scipy.ndimage import uniform_filter1d
```

```python
def Move_Avg(v, n):
    curve = uniform_filter1d(v, n)
    returnval = pd.Series(curve, index = v.index)
    return returnval
```

```python
kn = 10 # kernal length
```

```python
yfp_bg_s = yfp_bg.apply(Move_Avg, axis = 1, n = kn)
mkate_bg_s = mkate_bg.apply(Move_Avg, axis = 1, n = kn)
```
